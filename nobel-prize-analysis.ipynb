<B>The Nobel Prize</B> is a set of prestigious international awards presented annually in several categories, recognizing outstanding contributions in the fields of Physics, Chemistry, Medicine, Literature, Peace, and Economic Sciences. Established by the will of Alfred Nobel, a Swedish inventor, scientist, and philanthropist, in 1895, the prizes are awarded to individuals and organizations that have made significant advancements or contributions to humanity in these areas. The Nobel Prizes are widely regarded as some of the most esteemed honors, celebrating those who have made exceptional strides in promoting knowledge, peace, and positive global impact. The award ceremonies take place on December 10th each year, the anniversary of Alfred Nobel's death, and are accompanied by lectures and events that highlight the laureates' achievements.
    
Between 1901 and 2023, the Nobel Prizes and the Sveriges Riksbank Prize in Economic Sciences in Memory of Alfred Nobel were awarded 621 times to 1,000 people and organisations. With some receiving the Nobel Prize more than once, this makes a total of 965 individuals and 27 organisations. Below, we can view the full list of Nobel Prizes and Nobel Prize laureates. **Gender Distribution of Prize Winners**

```python
# Filter the DataFrame for 'male' and 'female' genders
gender_df = df[df['gender'].isin(['male', 'female'])]

# Set a color palette for the countplot
palette = "viridis"

# Countplot of Nobel Prize distribution by gender
plt.figure(figsize=(8, 4))
ax = sns.countplot(x='gender', data=gender_df, palette=palette)

# Display values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points',
                fontsize=12, color='black')  # Adjust font size and color

# Title and labels
plt.title('Gender Distribution of Nobel Prize Winners', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show the plot
plt.show()
```

**Nobel Prize Distribution Over Years**

```python
# Define a style dictionary
plot_style_line = {
    'color': 'darkblue',
    'marker': 'o',
    'markersize': 8,
    'linestyle': '-',
    'linewidth': 2,
    'title': {'text': 'Nobel Prize Distribution Over Years', 'fontsize': 16, 'color': 'darkblue'},
    'xlabel': {'text': 'Year', 'fontsize': 14, 'color': 'gray'},
    'ylabel': {'text': 'Prize Share', 'fontsize': 14, 'color': 'gray'},
    'grid': {'axis': 'both', 'linestyle': '--', 'alpha': 0.5, 'color': 'gray'}
}

# Line plot of Nobel Prize distribution over years with styling
plt.figure(figsize=(14, 6))
sns.lineplot(x='year', y='prizeShare', data=df, color=plot_style_line['color'], marker=plot_style_line['marker'],
             markersize=plot_style_line['markersize'], linestyle=plot_style_line['linestyle'],
             linewidth=plot_style_line['linewidth'])

# Title and labels
plt.title(plot_style_line['title']['text'], fontsize=plot_style_line['title']['fontsize'], color=plot_style_line['title']['color'])
plt.xlabel(plot_style_line['xlabel']['text'], fontsize=plot_style_line['xlabel']['fontsize'], color=plot_style_line['xlabel']['color'])
plt.ylabel(plot_style_line['ylabel']['text'], fontsize=plot_style_line['ylabel']['fontsize'], color=plot_style_line['ylabel']['color'])

# Add grid lines for better visibility
plt.grid(axis=plot_style_line['grid']['axis'], linestyle=plot_style_line['grid']['linestyle'],
         alpha=plot_style_line['grid']['alpha'], color=plot_style_line['grid']['color'])

# Show the plot
plt.show()
```

**Nobel Prize Distribution by Category**

```python
# Set a color palette for the countplot
palette = "viridis"

# Countplot of Nobel Prize distribution by category
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='category', data=df, order=df['category'].value_counts().index, palette=palette)

# Display values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Title and labels
plt.title('Nobel Prize Distribution by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Remove grid lines
ax.yaxis.grid(False)

# Show the plot
plt.show()
```

**Time Trends in Different Categories**

```python
# Set a color palette for the line plots
palette = "viridis"

# Line plots of Nobel Prize distribution over years for each category with styling
g = sns.FacetGrid(df, col='category', col_wrap=4, height=4, palette=palette)
g.map(sns.lineplot, 'year', 'prizeShare', linewidth=2)

# Set titles for each facet
g.set_titles(col_template="{col_name}", fontsize=16, color='darkblue')

# Set overall title
g.fig.suptitle('Time Trends in Nobel Prize Distribution by Category', y=1.02, fontsize=20, color='darkblue')

# Set axis labels
g.set_axis_labels('Year', 'Prize Share', fontsize=14, color='gray')

# Add legend for each facet
for ax in g.axes.flat:
    ax.legend(title='Category', title_fontsize='12', fontsize='10', loc='upper right')

# Add grid lines for better visibility
g.fig.tight_layout(rect=[0, 0, 1, 0.96])
g.fig.subplots_adjust(top=0.9)
g.despine(left=True, bottom=True)

# Show the plots
plt.show()
```

**Motivation Sentiment Analysis**

```python
from textblob import TextBlob  # Import TextBlob

# Set a color palette for the scatter plot
palette = "viridis"

# Sentiment analysis of motivation text
df['motivation_sentiment'] = df['motivation'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Scatter plot of motivation sentiment over years with styling
plt.figure(figsize=(14, 6))
sns.scatterplot(x='year', y='motivation_sentiment', data=df, palette=palette, alpha=0.7)

# Title and labels
plt.title('Motivation Sentiment Over Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Motivation Sentiment', fontsize=14)

# Add grid lines for better visibility
plt.grid(axis='both', linestyle='--', alpha=0.5, color='lightgray')

# Show the plot
plt.show()
```

**Poportion of USA Dominance Over Decades**

```python
# Calculating the proportion of USA born winners per decade
df['usa_born_winner'] = df['bornCountry'] == 'USA'
df['decade'] = (np.floor(df['year']/10) * 10).astype(int)
prop_usa_winners = df.groupby('decade', as_index=False)['usa_born_winner'].mean()

# Display the proportions of USA born winners per decade
prop_usa_winners.style.set_properties(**{"background-color": "#00A2A5","color":"white","border": "1.5px  solid black"})
```

```python
# Setting the plotting theme
sns.set()
# and setting the size of all plots
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(x='decade', y='usa_born_winner', data=prop_usa_winners, color='green', marker='o', markersize=8, linestyle='-', linewidth=2)

# Adding %-formatting to the y-axis
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))

# Adding labels and title
plt.xlabel('Decade', fontsize=14, color='gray')
plt.ylabel('Proportion of USA Born Winners', fontsize=14, color='gray')
plt.title('Proportion of USA Born Nobel Prize Winners Over Decades', fontsize=16, color='darkblue')

# Show the plot
plt.show()
```

**Age of Winners get their Prize**

```python
# Converting 'born' column to datetime, handling errors by setting invalid dates to NaT
df['born'] = pd.to_datetime(df['born'], errors='coerce', dayfirst=True)

# Dropping rows with NaT values in 'born' column
df = df.dropna(subset=['born'])

# Calculating the age of Nobel Prize winners
df['age'] = df['year'] - df['born'].dt.year

# Plotting the age of Nobel Prize winners
sns.lmplot(x='year', y='age', data=df, lowess=True, 
           aspect=2, line_kws={'color' : 'orange'})

# Adding the title
plt.title('Age of Nobel Prize Winners Over Time', fontsize=16, color='darkblue')

# Show the plot
plt.show()
```
