Unemployment is a critical economic indicator that reflects the labor market's health within a country. It not only impacts individuals but also influences the overall socioeconomic stability. Understanding global unemployment trends is essential for policymakers, economists, and researchers seeking insights into the dynamic nature of labor markets.


Explore the dynamics of global unemployment through a comprehensive dataset covering the years 2014 to 2024. This dataset includes Annual Unemployment Rates across various countries, emphasizing factors like gender and age groups. Join us in this brief yet insightful journey as we conduct an Exploratory Data Analysis (EDA) to uncover patterns and trends, contributing to a better understanding of the global employment landscape.

# <p style="padding:20px;background-color:#FD9799;margin:0;color:white;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 10px 10px;overflow:hidden;font-weight:500">EDA: Global Unemployment Rate</p>

```python
# Unemployment rate distribution in the year of 2024
sns.histplot(df['2024'], bins=20, kde=True)
plt.title('Unemployment Rate Distribution in 2024')
plt.xlabel('Unemployment Rate (%)')
plt.show()
```

**Exploring Relationships between Variables**

```python
sns.pairplot(df[['2020', '2021', '2022', '2023']])
plt.show()
```

**Exploring the Trends over Time**

```python
# Average unemployment rate by year
avg_unemployment_year = df[['2014', '2015', '2016', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']].mean()
plt.figure(figsize=(12,6))
avg_unemployment_year.plot(kind='line', marker='o')
plt.title('Average Unemployment Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate %')
plt.show()
```

**Exploring Relationship**

```python
# Boxplot of unemployment rate by gender
sns.boxplot(x='sex', y='2023', data=df)
plt.title('Unemployment Rate by Gender in 2023')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

**Gender and Age Group Analysis**

```python
# Compare unemployment rates across gender and age groups
plt.figure(figsize=(12,6))
sns.barplot(x='sex', y='2023', hue='age_group', data=df)
plt.title('Unemployment Rate by Gender and Age Group in 2023')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

**Comparative Analysis**

```python
# Compare unemployment rates between countries in a specific year
plt.figure(figsize=(20, 6))
sns.barplot(x='country_name', y='2023', data=df)
plt.title('Unemployment Rate Comparison Across Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate %')
plt.xticks(rotation=90)
plt.show()
```

**Correlation Analysis**

```python
# Exclude non-numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include='number').columns
correlation_matrix = df[numeric_columns].corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
