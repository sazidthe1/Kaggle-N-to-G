### Univariate Analysis

```python
# Set the style to white background
sns.set_style("white")

# Example of univariate analysis for numerical variable 'value'
plt.figure(figsize=(12, 6))
sns.histplot(df['value'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Value')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Univariate Analysis - Categorical Variable

```python
# Set the style to white background
sns.set_style("white")

# Example of univariate analysis for categorical variable 'country_name'
plt.figure(figsize=(12, 6))
sns.countplot(x='country_name', data=df, color='blue')
plt.title('Count of Data by Country')
plt.xticks(rotation=90)
plt.show()
```

### Bivariate Analysis

```python
# Set the style to white background
sns.set_style("white")

# Example of bivariate analysis between 'value' and 'country_name'
plt.figure(figsize=(12, 8))
sns.violinplot(x='country_name', y='value', data=df, color='blue')
plt.title('Value Distribution by Country')
plt.xticks(rotation=90)
plt.show()
```

```python
# Set the style to white background
sns.set_style("white")

# Example of bivariate analysis between 'value' and 'parameter'
plt.figure(figsize=(12, 6))
sns.boxplot(x='parameter', y='value', data=df, color='blue')
plt.title('Value by Parameter')
plt.xticks(rotation=45)
plt.show()
```

```python
# Set the style to white background
sns.set_style("white")

# Example of bivariate analysis between 'value' and 'parameter'
plt.figure(figsize=(10, 6))
sns.boxplot(x='product', y='value', data=df, color='blue')
plt.title('Value by Product')
plt.xticks(rotation=45)
plt.show()
```
