Global air pollution poses a significant environmental and public health challenge, resulting from various human activities and natural processes. Emissions from industries, transportation, agriculture, and energy production release pollutants such as particulate matter, nitrogen oxides, sulfur dioxide, and volatile organic compounds into the atmosphere. 

These pollutants contribute to smog formation, acid rain, climate change, and adverse health effects, impacting respiratory and cardiovascular systems and exacerbating existing health conditions. Mitigating global air pollution requires concerted efforts through policy interventions, technological innovations, and public awareness campaigns to safeguard human health and preserve the planet's ecosystems.

**Visualize AQI Distribution**

```python
# Set seaborn style and palette
sns.set_style('whitegrid')
sns.set_palette('magma')

# Plotting
plt.figure(figsize=(12, 6))
sns.histplot(df['aqi_value'], bins=20, kde=True, edgecolor='black')
plt.title('Distribution of AQI Values')
plt.xlabel('AQI Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

**Pollutant AQI Distribution**

```python
# Set seaborn style and palette
sns.set_style('whitegrid')
sns.set_palette('magma')

# Plotting
plt.figure(figsize=(12, 8))  # Larger figure size for better visibility
sns.boxplot(data=df[['aqi_value', 'co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']])
plt.title('Boxplot of AQI Values', fontsize=16)  # Larger font size for title
plt.ylabel('AQI Value', fontsize=14)  # Adjust font size for y-axis label
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Increase visibility of grid lines and ticks
plt.tick_params(axis='both', which='both', length=5, width=1, labelsize=12)  

plt.show()
```

**AQI Value - Top 50 Countries**

```python
# Set seaborn style to white
sns.set_style("white")

# Calculate average AQI value by country
avg_aqi_by_country = df.groupby('country_name')['aqi_value'].mean().sort_values(ascending=False)

# Select top 50 countries
top_50_countries = avg_aqi_by_country.head(50)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(x=top_50_countries.values, y=top_50_countries.index, palette='magma')
plt.title('Top 50 Countries by Average AQI Value')
plt.xlabel('Average AQI Value')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

**Correlation Analysis**

```python
# Calculate correlation matrix
correlation_matrix = df[['aqi_value', 'co_aqi_value\t', 'ozone_aqi_value', 'no2_aqi_value', 'pm2.5_aqi_value']].corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdPu', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
